```python
import pandas as pd
import numpy as np


data_train = pd.read_csv('C:/Users/alxro/Desktop/Courses/2IIG0/Assignments/HW1/heart_train_data.csv')
data_validate = pd.read_csv('C:/Users/alxro/Desktop/Courses/2IIG0/Assignments/HW1/heart_validate_data.csv')


data_validate

```

```python
from sklearn import preprocessing

array_chest_pain = data_train["Chest Pain"].to_numpy()
array_exang = data_train["Exang"].to_numpy()
array_thall = data_train["Thallium"].to_numpy()
array_target = data_train["Target"].to_numpy()

features = list(zip(array_chest_pain, array_exang, array_thall))
#print (features)

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(features, array_target)

predicted = model.predict([[1,1,2]])
print(predicted)
#print(array_target)
```

```python
# validate
fav = 0
tot = 0
test_arr = [0, 0, 0]
for i, row in data_validate.iterrows():
    st = row["Chest Pain"]
    nd = row["Exang"]
    rd = row["Thall"]
    test_arr[0] = st
    test_arr[1] = nd
    test_arr[2] = rd

    predicted = model.predict([test_arr])
    if predicted == row["Target"]:
        fav += 1
    tot += 1

print(fav, tot)
            
    
        
```

```python
# Learning
pyc = [0,0]

indexSet0 = []
indexSet1 = []

pxf_cp0 = [0,0]
pxf_cp1 = [0,0]
pxf_cp2 = [0,0]
pxf_cp3 = [0,0]

pxf_exang0 = [0,0]
pxf_exang1 = [0,0]

pxf_thal1 = [0,0]
pxf_thal2 = [0,0]
pxf_thal3 = [0,0]

for c in range (0, 2):
    for index, row in data_train.iterrows():
        if row['Target'] == c:
            pyc[c] += 1 / len(data_train["Target"])
    #print(pyc[c]
    print("pyc", c, " ", pyc[c])
  
    
for index, row in data_train.iterrows():
    if row['Target'] == 1:
        indexSet0.append(index)
    else:
        indexSet1.append(index)
        
for index, row in data_train.iterrows():
    if row["Chest Pain"] == 0:
        pxf_cp0[0] += 1 / len(indexSet0)
        pxf_cp0[1] += 1 / len(indexSet1) 
    if row["Chest Pain"] == 1:
        pxf_cp1[0] += 1 / len(indexSet0)
        pxf_cp1[1] += 1 / len(indexSet1) 
    if row["Chest Pain"] == 2:
        pxf_cp2[0] += 1 / len(indexSet0)
        pxf_cp2[1] += 1 / len(indexSet1) 
    if row["Chest Pain"] == 3:
        pxf_cp3[0] += 1 / len(indexSet0)
        pxf_cp3[1] += 1 / len(indexSet1) 
    
    if row["Exang"] == 0:
        pxf_exang0[0] += 1 / len(indexSet0)
        pxf_exang0[1] += 1 / len(indexSet1)
    if row["Exang"] == 1:
        pxf_exang1[0] += 1 / len(indexSet0)
        pxf_exang1[1] += 1 / len(indexSet1)
    
    if row["Thallium"] == 1:
        pxf_thal1[0] += 1 / len(indexSet0)
        pxf_thal1[1] += 1 / len(indexSet1)
    if row["Thallium"] == 2:
        pxf_thal2[0] += 1 / len(indexSet0)
        pxf_thal2[1] += 1 / len(indexSet1)
    if row["Thallium"] == 3:
        pxf_thal3[0] += 1 / len(indexSet0)
        pxf_thal3[1] += 1 / len(indexSet1)    
                


```
